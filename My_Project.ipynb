MOVIE RECOMMENDATION SYSTEM


Objective:To help the user choose a movie to watch.

```python
# Step 1: import library
import pandas as pd
```

```python
#Step 2: import data
film=pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Movies%20Recommendation.csv')
```

```python
# Data Decribe
film.head()
```

```python
film.info()
```

```python
film.describe()
```

```python
# Step 3: Define target(y) and features(X)
film.columns
```

```python
#Get the movie features
film_features=film[['Movie_Genre','Movie_Popularity','Movie_Keywords','Movie_Cast','Movie_Director']].fillna('')
```

```python
film_features.shape
```

```python
film_features
```

```python
X=film_features['Movie_Genre']+film_features['Movie_Keywords']+film_features['Movie_Cast']+film_features['Movie_Director']
```

```python
X.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
watch=TfidfVectorizer()
```

```python
X=watch.fit_transform(X)
```

```python
X.shape
```

```python
print(X)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(X)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
Favourite_movie_name=input('Enter your favourite movie name:')
```

```python
All_Movies_List=film['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movies_Recommendation=difflib.get_close_matches(Favourite_movie_name,All_Movies_List)
print(Movies_Recommendation)
```

```python
Close_Match=Movies_Recommendation[0]
print(Close_Match)
```

```python
Index_of_Close_Match_Movie=film[film.Movie_Title==Close_Match]['Movie_ID'].values[0]
print(Index_of_Close_Match_Movie)
```

```python
# Getting a List of similiar movies
Recommend_Score=list(enumerate(similarity_score[Index_of_Close_Match_Movie]))
print(Recommend_Score)
```

```python
len(Recommend_Score)
```

```python
# Sorting the movies based on the similarity_score
Sorted_Similar_Movies=sorted(Recommend_Score,key=lambda x:x[1],reverse=True)
print(Sorted_Similar_Movies)
```

```python
print('Top 20 Movies suggested for you: \n')
```

```python
i=1
for movie in Sorted_Similar_Movies:
  index=movie[0]
  title_from_index=film[film.index==index]['Movie_Title'].values[0]
  if(i<21):
    print(i,'.',title_from_index)
    i+=1
```

# Top 10 Movie Recommendation System

```python
Movie_Name=input('Enter your favourite movie name:')

list_of_all_movies=film['Movie_Title'].tolist()

Find_Close_Match=difflib.get_close_matches(Movie_Name,list_of_all_movies)

Close_Match=Find_Close_Match[0]

Index_of_Movie=film[film.Movie_Title==Close_Match]['Movie_ID'].values[0]

Recommendation_Score=list(enumerate(similarity_score[Index_of_Movie]))

Sorted_Similar_Movies=sorted(Recommendation_Score,key=lambda x:x[1],reverse=True)

print('Top 10 Movies suggested for you: \n')

i=1
for movie in Sorted_Similar_Movies:
  index=movie[0]
  title_from_index=film[film.index==index]['Movie_Title'].values[0]
  if(i<11):
    print(i,'.',title_from_index)
    i+=1
```
